```python
# DVS 128 Gesture Dataset
# https://www.research.ibm.com/dvsgesture/

import torch
import torch.nn as nn
from torch.autograd import Variable
from torch.nn.utils.rnn import pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader

class DVSGesture(Dataset):

  # Initialize your data, download, etc.
  def __init__(self, data_dir, is_train_set=False, transform=None):
    self.filenames = self.filename_list(is_train_set)
    self.records, self.labels = self.event_seq(self.filenames)
    self.gesture_mapping = self.gesture_list()
    self.len = len(self.labels)
    self.transform = transform

  def filename_list(self, is_train_set):
    file_dir = '/trials_to_train.txt' if is_train_set else '/trials_to_test.txt'

    with open(data_dir+file_dir, 'r') as f:
      filenames = f.readlines()

    return [data_dir + "/" + name[0:-1] for name in filenames]

  def event_seq(self,filenames):
    records = []
    labs = []

    for name in filenames:

      events, labels = aedat_to_events(name)
      for label in labels:
        i_start = np.searchsorted(events[:,0], label[1], side='left')
        i_end = np.searchsorted(events[:,0], label[2], side='left')
        event = events[i_start:i_end]
        records.append(event)
        labs.append(label[0])

    return records, labs

  def gesture_list(self):
    mapping = {}
    with open(data_dir + "/gesture_mapping.csv", 'r') as file:
      reader = csv.reader(file)
      for row in reader:
        if row[1] == 'label':
          continue
        mapping[row[1]] = row[0]

      return mapping

  def __getitem__(self, index):
    sample = (self.records[index], self.labels[index])
    if self.transform:
      sample = self.transform(sample)
    return sample

  def get_gesture_mapping(self):
    return self.gesture_mapping

  def get_gesture(self,id):
    return self.gesture_mapping[str(id)]

  def get_gesture_id(self,gesture_name):
    for id, gesture in self.gesture_mapping.items():
      if gesture == gesture_name:
        return id

  def __len__(self):
    return self.len
    
```

```python
class Flatten(object):
  def __init__(self, width, height, ds_factor):
    self.ds = pow(2,ds_factor)
    self.width = width
    self.height = height
    self.pol_off = (self.width//self.ds)*(self.height//self.ds)

  def __call__(self, sample):
    event_seq, label = sample
    flat_seq = []
    for event in event_seq:
      pos = (((self.width//self.ds)*(event[2]//self.ds))+(event[1]//self.ds))+(self.pol_off*event[3])
      flat_seq.append(pos)
    return np.array(flat_seq), label


# def eventSeqToNumArray(eventSeq,ds):
#   arr = [eventToNum(event,ds) for event in eventSeq]
#   return arr, len(arr)

# def eventToNum(event,ds):
#   width = 128;
#   height = 128;
#   ds = pow(2,ds)
#   offset = (width//ds)*(height//ds)*event[3] #width*height*polarization
#   return (((width//ds)*(event[2]//ds))+(event[1]//ds))+offset #width*ypos + xpos + offset
```

```python
def time_since(since):
  s = time.time() - since
  m = math.floor(s / 60)
  s -= m * 60
  return '%dm %ds' % (m, s)


def create_variable(tensor):
  # Do cuda() before wrapping with variable
  if torch.cuda.is_available():
      return Variable(tensor.cuda())
  else:
      return Variable(tensor)
```

```python
class RNNClassifier(nn.Module):
  # Our model

  def __init__(self, input_size, hidden_size, output_size, n_layers=1, bidirectional=True):
    super(RNNClassifier, self).__init__()
    self.hidden_size = hidden_size
    self.n_layers = n_layers
    self.n_directions = int(bidirectional) + 1

    self.embedding = nn.Embedding(input_size, hidden_size)
    self.gru = nn.GRU(hidden_size, hidden_size, n_layers,
                      bidirectional=bidirectional)
    self.fc = nn.Linear(hidden_size, output_size)

  def forward(self, input, seq_lengths):
    # Note: we run this all at once (over the whole input sequence)
    # input shape: B x S (input size)
    # transpose to make S(sequence) x B (batch)
    input = input.t()
    batch_size = input.size(1)

    # Make a hidden
    hidden = self._init_hidden(batch_size)

    # Embedding S x B -> S x B x I (embedding size)
    embedded = self.embedding(input)

    # Pack them up nicely
    gru_input = pack_padded_sequence(
        embedded, seq_lengths.data.cpu().numpy())

    # To compact weights again call flatten_parameters().
    self.gru.flatten_parameters()
    output, hidden = self.gru(gru_input, hidden)

    # Use the last layer output as FC's input
    # No need to unpack, since we are going to use hidden
    fc_output = self.fc(hidden[-1])
    return fc_output

  def _init_hidden(self, batch_size):
    hidden = torch.zeros(self.n_layers * self.n_directions,
                          batch_size, self.hidden_size)
    return create_variable(hidden)
```
